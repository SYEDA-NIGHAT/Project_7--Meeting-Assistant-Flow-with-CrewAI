```python
from crewai import Crew, Agent, Task

# Define an AI agent for meeting assistance
meeting_assistant = Agent(
    role="Meeting Assistant",
    goal="Summarize meetings and extract action items with assigned responsibilities.",
    backstory="A highly efficient AI trained to analyze meeting conversations and generate structured summaries and action points.",
    verbose=True  # Enables detailed logs if needed
)

# Define the tasks for the assistant
summarize_meeting = Task(
    description="Listen to the meeting conversation and generate a well-structured summary.",
    expected_output="A concise summary of the meeting with key points highlighted.",
    agent=meeting_assistant
)

extract_action_items = Task(
    description="Extract action items from the meeting and suggest who should take responsibility.",
    expected_output="A list of action items with assigned responsibilities.",
    agent=meeting_assistant
)

# Create the Crew with tasks
meeting_crew = Crew(
    agents=[meeting_assistant],
    tasks=[summarize_meeting, extract_action_items]
)

# Run the Crew to get results
results = meeting_crew.kickoff()

# Extract and print only the AI-generated responses (clean output)
formatted_results = "\n\n".join(task.raw for task in results.tasks_output if task.raw)

print("### Meeting Summary & Action Items ###\n")
print(formatted_results)
```
